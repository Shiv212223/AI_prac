```python
from queue import Queue

def water_jug_bfs(capacity_jug1, capacity_jug2, target):
    visited_states = set()
    q = Queue()

    # store (state, path)
    q.put(((0, 0), [(0, 0)]))

    while not q.empty():
        current_state, path = q.get()

        if current_state in visited_states:
            continue

        visited_states.add(current_state)

        jug1, jug2 = current_state

        # goal check
        if jug1 == target or jug2 == target:
            return path

        # Generate next states
        next_states = []

        # Fill jug1
        next_states.append((capacity_jug1, jug2))

        # Fill jug2
        next_states.append((jug1, capacity_jug2))

        # Empty jug1
        next_states.append((0, jug2))

        # Empty jug2
        next_states.append((jug1, 0))

        # Pour jug1 -> jug2
        pour = min(jug1, capacity_jug2 - jug2)
        next_states.append((jug1 - pour, jug2 + pour))

        # Pour jug2 -> jug1
        pour = min(jug2, capacity_jug1 - jug1)
        next_states.append((jug1 + pour, jug2 - pour))

        # Add to queue with updated path
        for state in next_states:
            if state not in visited_states:
                q.put((state, path + [state]))

    return None


result_path = water_jug_bfs(5, 6, 3)

if result_path:
    print("Water Jug Solution Path:")
    
    for step in result_path:
        print(step)
```
