```python
board = [" "] * 9

# Function to display board
def show():
    print(f"""
 {board[0]} | {board[1]} | {board[2]}
---|---|---
 {board[3]} | {board[4]} | {board[5]}
---|---|---
 {board[6]} | {board[7]} | {board[8]}
""")

# Winning combinations
win = [
    (0,1,2), (3,4,5), (6,7,8),
    (0,3,6), (1,4,7), (2,5,8),
    (0,4,8), (2,4,6)
]

player = "X"
game_over = False

# Main game loop
for _ in range(9):

    show()

    move = int(input(f"Player {player}, choose option (1-9): ")) - 1

    # Validate move
    if move < 0 or move > 8 or board[move] != " ":
        print("Invalid input")
        continue

    # Place move
    board[move] = player

    # Check winner
    for a, b, c in win:
        if board[a] == board[b] == board[c] == player:
            show()
            print(f"Player {player} wins!")
            game_over = True
            break

    # Stop game if won
    if game_over:
        break

    # Switch player
    player = "O" if player == "X" else "X"

# Draw condition
if not game_over:
    show()
    print("Draw")
```
