```python
# Travelling Salesperson Problem using Heuristic Search

def tsp_nearest_neighbor(graph, start):

    n = len(graph)

    visited = [False] * n

    path = [start]

    visited[start] = True

    total_cost = 0

    current_city = start

    for _ in range(n - 1):

        nearest_city = None

        min_distance = float('inf')

        for city in range(n):

            if not visited[city] and graph[current_city][city] < min_distance:

                min_distance = graph[current_city][city]

                nearest_city = city

        path.append(nearest_city)

        visited[nearest_city] = True

        total_cost += min_distance

        current_city = nearest_city

    total_cost += graph[current_city][start]

    path.append(start)

    return path, total_cost


# Distance Matrix
graph = [

    [0, 10, 15, 20],

    [10, 0, 35, 25],

    [15, 35, 0, 30],

    [20, 25, 30, 0]

]

start_city = 0

path, cost = tsp_nearest_neighbor(graph, start_city)

print("Shortest Path:", path)

print("Minimum Cost:", cost)
```
